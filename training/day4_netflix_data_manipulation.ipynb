```python
import pandas as pd

# Load Netflix dataset (must be uploaded in Colab or placed in local path)
df = pd.read_csv("netflix_titles.csv")
df.head()
```

## Column Overview & Data Types

```python

df.columns
df.dtypes
df.info()
```

## Filtering and Slicing

```python

# Filter shows from India
df[df['country'] == 'India']

# Filter Movies only
df[df['type'] == 'Movie']

# Filter by release year and rating
df[(df['release_year'] > 2015) & (df['rating'] == 'TV-MA')]
```

## Sorting

```python

# Sort by release year descending
df.sort_values('release_year', ascending=False).head()

# Sort by type then title
df.sort_values(by=['type', 'title']).head()
```

## Handling Missing Data

```python

# Check missing values
df.isnull().sum()

# Fill missing values with 'Unknown'
df.fillna('Unknown', inplace=True)
```

## GroupBy Operations

```python

# Count shows per country
df.groupby('country')['show_id'].count().reset_index().sort_values('show_id', ascending=False).head()

# Count by type and rating
df.groupby(['type', 'rating'])['show_id'].count().reset_index().sort_values('show_id', ascending=False).head()
```

## Merging Demo DataFrames

```python

# Simulated merge example
platform_data = pd.DataFrame({
    'platform': ['Netflix', 'Netflix', 'Netflix'],
    'country': ['India', 'United States', 'Canada'],
    'subscribers_million': [22, 73, 12]
})

country_counts = df.groupby('country')['show_id'].count().reset_index().rename(columns={'show_id': 'content_count'})

# Merge platform stats with content count
merged = pd.merge(platform_data, country_counts, on='country', how='left')
merged
```
